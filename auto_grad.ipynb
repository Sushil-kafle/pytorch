```python
import torch
import matplotlib.pyplot as plt
import math
```

```python
a = torch.linspace(0.,1.*math.pi,steps=25,requires_grad=True)
```

```python
a
```

```python
b = torch.sin(a)
with torch.no_grad():
    plt.plot(a,b)
```

```python
b.backward(torch.ones(a.shape))
```

```python
a.grad
```

```python
with torch.no_grad():
    plt.plot(a,b)
    plt.plot(a,a.grad)
plt.show()
```
